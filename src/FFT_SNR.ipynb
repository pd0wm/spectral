```python
import cogradio as cg
import bokeh.plotting
%matplotlib inline
#mpld3.enable_notebook()
bokeh.plotting.output_notebook()
import matplotlib.pyplot as plt
```

```python
frequencies = [6, 20]
widths = [1, 5]
f_s = 1e3
window = 4
snr = 20



signal = cg.source.Rect(frequencies, widths, SNR=snr).generate(f_s, window)
signal_fft = cg.fft(signal)

plt.plot(cg.freq_axis(f_s, window), signal_fft)
plt.show()

plot = bokeh.plotting.figure(title="FFT", 
                          x_axis_label='f',
                          y_axis_label='|x|',
                         plot_height=300)
plot.line(cg.freq_axis(f_s, window), 
          signal_fft, 
          line_width=2)
bokeh.plotting.show(plot)

plot2 = bokeh.plotting.figure(title="Sinusoidal", 
                           x_axis_label='t', 
                           y_axis_label='x',
                          plot_height=300)
plot2.line(cg.time_axis(f_s, window), 
           signal, 
           line_width=2)
bokeh.plotting.show(plot2)
```
